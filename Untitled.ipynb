```python
def get_popular(Series):
    langauges = []
    for i in Series:
        if pd.isnull(i):
            pass
        else:
            split = str(i).split(";")
            _ = [langauges.append(word) for word in split]
        return pd.Series(langauges).value_counts()
        
def return_top_n(Series, n):
    Series = get_popular(Series)
    return Series.head(n)

def convert_to_yearly_comp(CompTotal, CompFreq):
    if CompFreq == 'Monthly':
        return CompTotal * 12
    elif CompFreq == 'Weekly':
        return CompTotal * 52
    else:
        return CompTotal
    
def language_var(df, col, language, prefix = "user"):
    df[language + '_' + prefix] = [1 if language in str(row) else 0 for row in df[col]]
    return df
    
def fix_comp(df):
    df['CompTotal'] = [convert_to_yearly_comp(CompTotal, CompFreq) for CompTotal, CompFreq in zip(df['CompTotal'],
                                                                                                 df['CompFreq'])]
    return df
```

```python
def fix_1st_code(row): 
    if row == 'Younger than 5 years':
        return 4
    elif row == 'Older than 85':
        return 86
    else:
        return int(row)
    
model_frame['Age1stCode'] = model_frame['Age1stCode'].apply(fix_1st_code)
model_frame['YearsCoding'] = model_frame['Age'] - model_frame['Age1stCode']
```

```python
def ordinal_org_size(row):
    if row == 'Just me - I am a freelancer, sole proprietor, etc.':
        return 1
    elif row == '2 to 9 employees':
        return 2
    elif row == '10 to 19 employees':
        return 3
    elif row == '20 to 99 employees':
        return 4
    elif row == '100 to 499 employees':
        return 5
    elif row == '500 to 999 employees':
        return 6
    elif row == '1,000 to 4,999 employees':
        return 7
    elif row == '5,000 to 9,999 employees':
        return 8
    else:
        return 9
    
model_frame['OrgSize'] = model_frame['OrgSize'].apply(ordinal_org_size)
```

```python
def ordinal_ed(row):
    if row == 'I never completed any formal education':
        return 1
    elif row == 'Primary/elementary school':
        return 2
    elif row == 'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)':
        return 3
    elif row == 'Some college/university study without earning a degree':
        return 4
    elif row == 'Some college/university study without earning a degree':
        return 5
    elif row == 'Associate degree (A.A., A.S., etc.)':
        return 6
    elif row == 'Bachelor’s degree (B.A., B.S., B.Eng., etc.':
        return 7
    elif row == 'Bachelor’s degree (B.A., B.S., B.Eng., etc.':
        return 8
    elif row == 'Master’s degree (M.A., M.S., M.Eng., MBA, etc.)':
        return 9
    else:
        return 10
    
model_frame['OrgSize'] = [ordinal_ed(row) for row in model_frame['OrgSize']]
```

```python
def word_count(row):
    if pd.isnull(row):
        pass
    else:
        split = str(row).split(";")
        return split
    
    
dev_counts = pd.concat([pd.Series(word_count(row)) for row in df['DevType'].dropna()], 0).value_counts()
unique_types = dev_counts.index.unique()

misc_counts = pd.concat([pd.Series(word_count(row)) for row in df['MiscTechWorkedWith'].dropna()], 0).value_counts()
unique_misc = misc_counts.index.unique()
misc_counts.head()
unique_misc
```

```python
for i in unique_types:
    string = i
    model_frame[i] = [1 if row == i else 0 for row in model_frame['DevType']]
    
for i in unique_misc:
    string = i
    model_frame[i] = [1 if row == i else 0 for row in model_frame['MiscTechWorkedWith']]
```

```python
from sklearn.model_selection import train_test_split

model_frame = model_frame.dropna().drop(['DevType'], 1)
catgor_cols = []
for i in model_frame.columns:
    if model_frame[i].dtype == 'object':
        catgor_cols.append(i)
    else:
        pass

y = model_frame['ConvertedComp']
X = model_frame.drop(['ConvertedComp'], 1)

X_train, X_test, y_train, y_test = train_test_split(X, y)

from catboost import CatBoostRegressor

learner = CatBoostRegressor(cat_features = catgor_cols, n_estimators = 1000, silent = True, learning_rate = 0.1)
learner = learner.fit(X_train, y_train)
learner.score(X_test, y_test)
```

```python
import shap
explainer = shap.TreeExplainer(learner)
shap_values = explainer.shap_values(X_train)
shap.summary_plot(shap_values, X_train)
```

```python
shap.dependence_plot("OpSys", shap_values, X_train)
```
